```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## line plot

```python
x=[1,2,3,4,5]
y=[10,20,40,50,40]
plt.plot(x,y);
plt.title("simple line plot");
plt.xlabel('X-axis Lablel');
plt.ylabel('Y-axis Label');
plt.grid(True)
```

## Bar Chart

```python
months=['Jan','Feb','Mar','April','May','June','July','Aug','Sep','Oct','Nov','Dec']
rev=[100,200,400,300,500,600,700,800,900,1000,1100,1200]
```

```python
plt.figure(figsize=(10,6));
plt.bar(months,rev)
plt.xlabel('Months');
plt.ylabel("rev");
```

## Scatter Plot

```python
heights=[150,170,155,180]
weights=[85,78,76,58]
plt.scatter(heights,weights);
plt.grid()
plt.title("height vs weight")
plt.xlabel("heights");
plt.ylabel("weights");
```

## subplots and Legends

```python
days=[1,2,3,4,5]
original_price=[20,50,40,30,45];
predicted_price=[30,70,45,30,50];
plt.plot(days,original_price,'b-o', label='original_price');
plt.plot(days,predicted_price,'r-o',label='predicted_price');
plt.legend();
plt.title("Stock Price");
plt.xlabel("No of days");
plt.ylabel('predicted vs origanl prices');
```

## subplots and Legends 

```python
days=[1,2,3,4,5]
```

```python
plt.subplot(3,1,1);
plt.plot(days,original_price,'b-o', label='original_price');
plt.grid();
plt.legend();
plt.title('Original Price')

```

```python
plt.subplot(3,1,2);
plt.plot(days,predicted_price,'r-o',label='predicted_price');
plt.legend();
plt.title('Predicted price');

```

```python
# price difference
plt.subplot(3,1,3);
p_diff = [o - p for o, p in zip(original_price, predicted_price)]
plt.bar(days,original_price,color='purple', label='original_price');
plt.bar(days,predicted_price,color='pink',label='predicted_price');

```

```python
plt.subplot(3,1,2);
plt.plot(days,predicted_price,'r-o',label='predicted_price');
plt.legend();
plt.title('Predicted price');
```
