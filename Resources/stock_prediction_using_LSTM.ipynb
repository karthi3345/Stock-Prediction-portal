```python
# =========================
# 1. Imports
# =========================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input

# =========================
# 2. Load dataset
# =========================
df = pd.read_csv("TSLA.csv")   # <-- change file name if needed

# =========================
# 3. Take Close price
# =========================
close_price = df['Close'].values

# =========================
# 4. Scale data (ONLY ONCE)
# =========================
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(close_price.reshape(-1,1))

# =========================
# 5. Train / Test split
# =========================
train_size = int(len(scaled_data) * 0.8)

train_data = scaled_data[:train_size]
test_data  = scaled_data[train_size:]

# =========================
# 6. Create sequences
# =========================
def create_sequences(data, lookback=100):
    X, y = [], []
    for i in range(lookback, len(data)):
        X.append(data[i-lookback:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

x_train, y_train = create_sequences(train_data)
x_test, y_test   = create_sequences(test_data)

# =========================
# 7. Reshape for LSTM
# =========================
x_train = x_train.reshape((x_train.shape[0], x_train.shape[1], 1))
x_test  = x_test.reshape((x_test.shape[0], x_test.shape[1], 1))

# =========================
# 8. Build model
# =========================
model = Sequential()
model.add(Input(shape=(100,1)))
model.add(LSTM(128, return_sequences=True))
model.add(LSTM(64))
model.add(Dense(25))
model.add(Dense(1))

# =========================
# 9. Compile & Train
# =========================
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, epochs=50, batch_size=32)

# =========================
# 10. Predict
# =========================
y_predicted = model.predict(x_test)

# =========================
# 11. Inverse scaling
# =========================
y_predicted = scaler.inverse_transform(y_predicted)
y_test = scaler.inverse_transform(y_test.reshape(-1,1))

# =========================
# 12. Plot result
# =========================
plt.figure(figsize=(12,6))
plt.plot(y_test, label='Actual', color='blue')
plt.plot(y_predicted, label='Predicted', color='red')
plt.title("Stock Price Prediction")
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.show()

# =========================
# 13. Sanity check
# =========================
print("Close price range:", close_price.min(), close_price.max())
print("y_test range:", y_test.min(), y_test.max())
print("y_pred range:", y_predicted.min(), y_predicted.max())

```
